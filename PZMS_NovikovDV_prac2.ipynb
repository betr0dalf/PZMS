```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import scipy.stats as stats

# Загрузка данных

data = pd.read_csv('1.txt', sep='\t', header=None, names=['number'])
data_2 = pd.read_csv('2.txt', sep='\t', header=None, names=['number'])
data_3 = pd.read_csv('3.txt', sep='\t', header=None, names=['number'])
data_4 = pd.read_csv('4.txt', sep='\t', header=None, names=['number'])

# Задание 1

# Считаем среднее значение и среднеквадратичное отклонение

mean_num_1 = statistics.mean(data['number'])
std_dev_1 = statistics.pstdev(data['number'])

mean_num_2 = statistics.mean(data_2['number'])
std_dev_2 = statistics.pstdev(data_2['number'])

mean_num_3 = statistics.mean(data_3['number'])
std_dev_3 = statistics.pstdev(data_3['number'])

mean_num_4 = statistics.mean(data_4['number'])
std_dev_4 = statistics.pstdev(data_4['number'])

# Задание 2

# confidence_interval_21
n_1 = len(data)
confidence_level_1 = 0.95 # Уровень доверия
# Критическое значение z функции Лапласа для заданного уровня доверия
z_1 = stats.norm.ppf((1 + confidence_level_1) / 2)
# Вычисление доверительного интервала (убираем выбросы)
margin_of_error_1 = z_1 * (std_dev_1 / np.sqrt(n_1))
confidence_interval_21 = (round(mean_num_1 - margin_of_error_1, 3), round(mean_num_1 + margin_of_error_1, 3))

# confidence_interval_22
n_2 = len(data_2)
confidence_level_2 = 0.95
z_2 = stats.norm.ppf((1 + confidence_level_2) / 2)
margin_of_error_2 = z_2 * (std_dev_2 / np.sqrt(n_2))
confidence_interval_22 = (round(mean_num_2 - margin_of_error_2, 3), round(mean_num_2 + margin_of_error_2, 3))

# confidence_interval_23
n_3 = len(data_3)
confidence_level_3 = 0.95
z_3 = stats.norm.ppf((1 + confidence_level_3) / 2)
margin_of_error_3 = z_3 * (std_dev_3 / np.sqrt(n_3))
confidence_interval_23 = (round(mean_num_3 - margin_of_error_3, 3), round(mean_num_3 + margin_of_error_3, 3))

# confidence_interval_24
n_4 = len(data_4)
confidence_level_4 = 0.95
z_4 = stats.norm.ppf((1 + confidence_level_4) / 2)
margin_of_error_4 = z_4 * (std_dev_4 / np.sqrt(n_4))
confidence_interval_24 = (round(mean_num_4 - margin_of_error_4, 3), round(mean_num_4 + margin_of_error_4, 3))

# Задание 3

# t-распределение Стьюдента лучше работает на маленьких выборках для определения доверительного интервала

# confidence_interval_31
n_31 = len(data)
confidence_level_31 = 0.95
degrees_of_freedom_31 = n_31 - 1
# Критическое значение t для заданного уровня доверия и степеней свободы
t_critical_31 = stats.t.ppf((1 + confidence_level_31) / 2, degrees_of_freedom_31)
# Вычисление доверительного интервала, компенсируем смещенность
margin_of_error_31 = t_critical_31 * (std_dev_1 / np.sqrt(n_31))
confidence_interval_31 = (round(mean_num_1 - margin_of_error_31, 3), round(mean_num_1 + margin_of_error_31, 3))

# confidence_interval_32
n_32 = len(data_2)
confidence_level_32 = 0.95
degrees_of_freedom_32 = n_32 - 1
t_critical_32 = stats.t.ppf((1 + confidence_level_32) / 2, degrees_of_freedom_32)
margin_of_error_32 = t_critical_32 * (std_dev_2 / np.sqrt(n_32))
confidence_interval_32 = (round(mean_num_2 - margin_of_error_32, 3), round(mean_num_2 + margin_of_error_32, 3))

# confidence_interval_33
n_33 = len(data_3)
confidence_level_33 = 0.95
degrees_of_freedom_33 = n_33 - 1
t_critical_33 = stats.t.ppf((1 + confidence_level_33) / 2, degrees_of_freedom_33)
margin_of_error_33 = t_critical_33 * (std_dev_3 / np.sqrt(n_33))
confidence_interval_33 = (round(mean_num_3 - margin_of_error_33, 3), round(mean_num_3 + margin_of_error_33, 3))

# confidence_interval_34
n_34 = len(data_4)
confidence_level_34 = 0.95
degrees_of_freedom_34 = n_34 - 1
t_critical_34 = stats.t.ppf((1 + confidence_level_34) / 2, degrees_of_freedom_34)
margin_of_error_34 = t_critical_34 * (std_dev_4 / np.sqrt(n_34))
confidence_interval_34 = (round(mean_num_4 - margin_of_error_34, 3), round(mean_num_4 + margin_of_error_34, 3))

# Задание 4

def Square(data, mean_num, std_dev):
   # Вычисление выборочной дисперсии
    sample_variance = statistics.variance(data['number'])
    n = len(data)

    # Уровень доверия (например, 95%)
    confidence_level = 0.95
    alpha = 1 - confidence_level

    # Степени свободы
    degrees_of_freedom = n - 1

    # Критические значения χ² для заданного уровня доверия и степеней свободы
    chi2_lower = stats.chi2.ppf(alpha / 2, degrees_of_freedom)
    chi2_upper = stats.chi2.ppf(1 - alpha / 2, degrees_of_freedom)

    # Вычисление доверительного интервала для стандартного отклонения
    confidence_interval_lower = np.sqrt((degrees_of_freedom * sample_variance) / chi2_upper)
    confidence_interval_upper = np.sqrt((degrees_of_freedom * sample_variance) / chi2_lower)

    return (round(confidence_interval_lower, 3), round(confidence_interval_upper, 3))

confidence_interval_41 = Square(data, mean_num_1, std_dev_1)
confidence_interval_42 = Square(data_2, mean_num_2, std_dev_2)
confidence_interval_43 = Square(data_3, mean_num_3, std_dev_3)
confidence_interval_44 = Square(data_4, mean_num_4, std_dev_4)


df = pd.DataFrame({
    '1.txt': [len(data), mean_num_1, std_dev_1, confidence_interval_21, confidence_interval_31, confidence_interval_41],
    '2.txt': [len(data_2), mean_num_2, std_dev_2, confidence_interval_22, confidence_interval_32, confidence_interval_42],
    '3.txt': [len(data_3), mean_num_3, std_dev_3, confidence_interval_23, confidence_interval_33, confidence_interval_43],
    '4.txt': [len(data_4), mean_num_4, std_dev_4, confidence_interval_24, confidence_interval_34, confidence_interval_44],
}, index=['Количество элементов', 'Среднее значение', 'Среднеквадратическое отклонение', 'Доверительный интервала математического ожидания по правилу нормального распределения', 't-распределения Стьюдента', 'Доверительный интервал для среднеквадратического отклонения'])

display(df)
```
